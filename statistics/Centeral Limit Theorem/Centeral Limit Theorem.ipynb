# Central Limit Theorem (CLT)

The **central limit theorem (CLT)** states that if you take a sufficiently large sample from a population, the distribution of the sample means will be approximately normally distributed, even if the population isn't normally distributed. The theorem is important because it allows us to use the normal distribution to make inferences about populations, even when we don't have data for the entire population.
